```python
class Dog:
    def __init__(self, name, gender, breed, age, weight, color):
        self.name = name
        self.gender = gender
        self.breed = breed

        # private attribute
        self.__age = age
        self.weight = weight
        self.color = color

    # setter
    def setAge(self, newAge):
        if newAge < self.__age:
            print("Error !! อายุใหม่ต้องมากกว่าเดิม")
        else:
            self.__age = newAge

    # getter
    def getAge(self):
        return self.__age

    def showInfo(self):
        print(f"""
ชื่อ: {self.name}
เพศ: {self.gender}
สายพันธุ์: {self.breed}
อายุ: {self.__age}
น้ำหนัก: {self.weight}
สี: {self.color}
""")

    def sit(self):
        print(f"{self.name} is sitting !!")

    def sleep(self):
        print(f"{self.name} is sleeping !!")

    def eat(self):
        print(f"{self.name} is eating !!")

```

```python
# Instance Variable / object
dog_ker = Dog("เกอร์", "เมีย", "ชิวาวา", 4, 20, "ดำ")
dog_war = Dog("วาร์", "ผู้", "บลูด็อก", 4, 20, "ขาว")
```

```python
dog_ker.showInfo()
```

```python
dog_war.showInfo()
```

```python
dog_ker.age = 3
```

```python
dog_ker.showInfo()
```
