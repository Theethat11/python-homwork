## ให้สร้าง class student สําหรับจำลองนักศึกษา
-รหัสนักศึกษา
-หมายเลขบัตร 13 หลัก
-ชื่อ
-นามสกุล
-อายุ
-เพศ


```python
class Student:
    def __init__(self, sid, cid, name, lname, age, gender):
        self.__sid = sid
        self.__name = name
        self.__lname = lname
        self.__age = age
        self.__gender = gender

        if len(cid) == 13:
            self.__cid = cid
        else:
            print("หมายเลขบัตรต้องมีความยาว 13 หลักเท่านั้น")
            self.__cid = ""

    def setAge(self, newAge):
        if self.__age < newAge:
            self.__age = newAge
        else:
            print("อายุต้องเพิ่มขึ้นมากกว่าเดิมเท่านั้น")

    def setCid(self, cid):
        if len(cid) == 13 and self.__cid == "":
            self.__cid = cid
        else:
            print("หมายเลขบัตรต้องมีความยาว 13 หลักเท่านั้น")

    def showInfo(self):
        print("""
ชื่อ: {}
นามสกุล: {}
รหัสนักศึกษา: {}
หมายเลขบัตรประชาชน: {}
อายุ: {}
""".format(
            self.__name,
            self.__lname,
            self.__sid,
            self.__cid,
            self.__age
        ))
```

```python
theethat = Student(
    '6812732111',
    '1339100050721',
    'ธีธัช',
    'สมพร',
    19,
    'ชาย'
)
```

```python
theethat.showInfo()
```

```python
theethat.showInfo()
```

```python
theethat.setAge(20)
theethat.showInfo()
```

```python
theethat.setAge(18)
```
