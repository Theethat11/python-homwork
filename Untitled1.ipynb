```python
class Person:
    def __init__(self, cid, name, lname):
        self.cid = cid
        self.name = name
        self.lname = lname
        
    def info(self):
        print ("{}: {} {}".format(self.cid, self.name, self.lname))
```

```python
class Student(Person):
    def __init__(self, cid, name, lname):
        super().__init__(cid, name, lname)

    def helloTeahcher(self):
        print("สวัสดีครับคุณครู")
```

```python
person1 = Person("1223334444555", "Theethat", "Somporn")
person1.info()
```

```python
student = Student("1223334444555", "Theethat", "Somporn")
student.info()
student.helloTeahcher()
```
