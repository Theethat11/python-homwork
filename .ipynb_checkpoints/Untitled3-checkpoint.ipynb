```python
'''
Method Override
'''

class Parent:
    def __init__(self):
        pass
        
    def hello(self):
        print("หวัดดี")
```

```python
class Child(Parent):
    def __init__(self):
        pass

    def hello(self):
        print("สวัสดีครับ")
```

```python
person = Child()
person.hello()
```
