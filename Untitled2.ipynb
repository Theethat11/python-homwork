```python
from datetime import datetime

class DigitClock:
    def __init__(self):
        pass

    def showTime(self):
        now = datetime.now()
        current_time = now.strftime("%H:%M:%S")
        print("เวลาปัจจุบัน : {}".format(current_time))
```
